```python
def Generate(t1):
    begin = "<!DOCTYPE html>\n<html>\n<body>\n"+t1+"\n</body>\n</html>"
    return begin

def div(head, body):
    return "<div>"+head+"\n"+body+ "\n</div>"

def H1(content):
    return "<h1>"+content+"</h1>"

def p(content):
    return "<p>"+content+"</p>"

Out = Generate(div(H1("This is my heading"), p("This is my paragraph")))
print(Out)
```
